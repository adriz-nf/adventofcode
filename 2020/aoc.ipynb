```python
# Day 8
with open('day8input.txt') as file:
    data = file.readlines()
    data = [line.strip().split() for line in data]

for delta_line in range(len(data)):
    accumulator = 0
    used_line_numbers = []
    line_number = 0
    results = []
    while line_number not in used_line_numbers and line_number<len(data):
        process_row = data[line_number]
        operation = process_row[0]
        value = int(process_row[1])
        used_line_numbers.append(line_number)
        if line_number == delta_line:
            if operation == 'nop':
                operation = 'jmp'
            elif operation == 'jmp':
                operation = 'nop'
                    
        if operation == 'nop':
            line_number += 1
        elif operation == 'acc':
            accumulator += value
            line_number += 1
        elif operation == 'jmp':
            line_number += value
    if line_number==len(data):
        print(delta_line, accumulator )
        break

    
```
